```python
# substituindo vírgula por ponto e convertendo coluna de texto para número
df_viagens['Valor diárias'] = df_viagens['Valor diárias'].str.replace(",", ".").astype(float)
```

```python
df_viagens['Valor diárias']
```

```python
# convertendo colunas de texto para número
df_viagens['Valor passagens'] = df_viagens['Valor passagens'].str.replace(",", ".").astype(float)
df_viagens['Valor devolução'] = df_viagens['Valor devolução'].str.replace(",", ".").astype(float)
df_viagens['Valor outros gastos'] = df_viagens['Valor outros gastos'].str.replace(",", ".").astype(float)
```

```python
# soma de despesas
df_viagens["Despesas"] = df_viagens["Valor diárias"] + df_viagens["Valor passagens"] + df_viagens["Valor outros gastos"]
```

# Lidando com dados nulos

```python
df_viagens
```

```python
# remoção de valores nulos em 'Cargo'
df_viagens.dropna(subset = 'Cargo')
```

```python
# substituindo 'NaN' por outra coisa
df_viagens['Cargo'] = df_viagens['Cargo'].fillna("NÃO IDENTIFICADO")
df_viagens
```

```python
df_viagens[["Período - Data de início", "Período - Data de fim"]]
```

```python
# conversão de object para datetime
pd.to_datetime(df_viagens['Período - Data de início'], format = "%d/%m/%Y")
```

```python
df_viagens['Período - Data de início'] = pd.to_datetime(df_viagens['Período - Data de início'], format = "%d/%m/%Y")
df_viagens['Período - Data de fim'] = pd.to_datetime(df_viagens['Período - Data de fim'], format = "%d/%m/%Y")
```

```python
df_viagens
```

```python
# mês
df_viagens['Mês da viagem'] = df_viagens['Período - Data de início'].dt.month_name()
```

```python
# intervalo de duração da viagem
df_viagens['Dias de duração da viagem'] = (df_viagens['Período - Data de fim'] - df_viagens['Período - Data de início']).dt.days
```

```python
df_viagens
```

```python
# excluindo uma coluna específica
#df_viagens = df_viagens.drop('Dias de viagem', axis = 1)
```

```python
# usando agregação de colunas e funções
(
    df_viagens
    .groupby('Cargo')
    .agg(
        despesa_media = ('Despesas', 'mean'),
        duracao_media = ('Dias de duração da viagem', 'mean'),
        despesas_totais = ('Despesas', 'sum'),
        destino_mais_frequente = ('Destinos', pd.Series.mode),
        n_viagens = ('Nome', 'count')
        )
    .reset_index()
    )
```

```python
# percentual de cargos
df_cargos = df_viagens['Cargo'].value_counts(normalize = True).reset_index()
df_cargos
```

```python
# filtro com loc
df_cargos.loc[df_cargos['proportion'] > 0.01, 'Cargo']
```

```python
df_viagens_consolidado = (
    df_viagens
    .groupby('Cargo')
    .agg(
        despesa_media = ('Despesas', 'mean'),
        duracao_media = ('Dias de duração da viagem', 'mean'),
        despesas_totais = ('Despesas', 'sum'),
        destino_mais_frequente = ('Destinos', pd.Series.mode),
        n_viagens = ('Nome', 'count')
        )
    .reset_index()
    )
```

```python
cargos_relevantes = df_cargos.loc[df_cargos['proportion'] > 0.01, 'Cargo']
```

```python
# cargos que representam pelo menos 1% das viagens
cargos_relevantes
```

```python
filtro = df_viagens_consolidado['Cargo'].isin(cargos_relevantes)
```

```python
df_final = df_viagens_consolidado[filtro]
```

```python
df_final
```

```python
df_final = df_final.sort_values(by = 'n_viagens', ascending = False)
```

```python
# criando gráficos
df_final.plot(x = 'Cargo', y = 'n_viagens', kind = 'bar')
```

```python
# importando a biblioteca matplotlib
import matplotlib.pyplot as plt
```

```python
# dimensionamento
fig, ax = plt.subplots(figsize = (16, 6))

# gráfico de barras horizontais
ax.barh(df_final['Cargo'], df_final['n_viagens'], color = '#91b1e6')

# ordena do maior para o menor valor
ax.invert_yaxis()

# alterar cor de fundo
ax.set_facecolor('#e6e8ed')

# título
fig.suptitle('Viagens por cargo público (2023)')

# texto de fonte dos dados
plt.figtext(0.78, 0.89,'Fonte: Portal da Transparência', fontsize = 8)

# estilização do fundo
plt.grid(color = '#d1dbed', linestyle = '--', linewidth = 0.5)

# legendas no eixo y
plt.yticks(fontsize = 9)

# legendando o eixo x
plt.xlabel('Número de viagens')

plt.show()
```
